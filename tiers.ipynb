```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import json

sns.set_style('darkgrid')
```

```python
r = requests.get('https://ronin.rest/origin/leaderboardStatistics')
df = pd.DataFrame.from_dict(r.json(), orient='index', columns=['accounts'])
df = df.groupby(df.index)['accounts'].sum().reset_index(name='accounts')
```

```python
order1 = ['Egg 4','Egg 3','Egg 2','Egg 1','Chick 4','Chick 3','Chick 2','Chick 1','Hare 4','Hare 3','Hare 2','Hare 1','Boar 4','Boar 3','Boar 2','Boar 1']
plt.figure(figsize=(14, 4))
bar = sns.barplot(data=df, x='index', y='accounts', estimator=sum, order=order1)
bar.set(title='ranking accounts - segment 1', ylabel='accounts', xlabel='rank')
bar.bar_label(bar.containers[0])

plt.show()
```

```python
order2 = ['Wolf 4','Wolf 3','Wolf 2','Wolf 1','Bear 4','Bear 3','Bear 2','Bear 1','Tiger 4','Tiger 3','Tiger 2','Tiger 1','Dragon 4','Dragon 3','Dragon 2','Dragon 1','Challenger']
plt.figure(figsize=(14, 4))
bar = sns.barplot(data=df, x='index', y='accounts', estimator=sum, order=order2)
bar.set(title='ranking accounts - segment 2', ylabel='accounts', xlabel='rank')
bar.bar_label(bar.containers[0])

plt.show()
```
